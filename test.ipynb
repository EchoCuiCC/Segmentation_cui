```python
import torch
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('dark')
```

```python
pub = torch.load('/root/workspace/work/nnUNetFrame/public_information.pth').cpu().cpu().detach().numpy().astype(np.float64)[1]
cen = torch.load('/root/workspace/work/nnUNetFrame/cen_feature.pth').cpu().detach().numpy().astype(np.float64)[1]
seg = torch.load('/root/workspace/work/nnUNetFrame/seg_feature.pth').cpu().detach().numpy().astype(np.float64)[1]
input = torch.load('/root/workspace/work/nnUNetFrame/inputs.pth').cpu().detach().numpy().astype(np.float64)[1]
```

```python
features = [input,seg,cen,pub]
name = ['input','seg','cen','pub']
```

```python

plt.figure(figsize=(40,10))
page = 51

potition = 1
for i in range(len(name)):
    for j in range(0,32,2):
        plt.subplot(len(name),16,potition)
        if i==0:
            plt.imshow(features[i][0,page])
        else:
            plt.imshow(features[i][j,page])
        plt.title(name[i])
        potition+=1
    plt.colorbar()
```
